```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
laps_2024 = pd.read_csv('../data/merged/laps_2024.csv')
```

```python
laps_2024.head()
```

```python
valid_laps = laps_2024[laps_2024['LapTime'].notnull()]
```

```python
valid_laps.shape
```

```python
laps_2024.shape
```

```python
valid_laps['LapTime'] = pd.to_timedelta(valid_laps['LapTime']).dt.total_seconds()
```

```python
valid_laps['LapTime']
```

```python
fastest_laps = valid_laps.groupby('Driver')['LapTime'].min().sort_values()
```

```python
valid_laps[['Driver', 'LapTime', 'RaceName']].sort_values('LapTime').head(20)
```

```python
import seaborn as sns
```

```python
fastest_laps_sorted = valid_laps.sort_values('LapTime')
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=fastest_laps_sorted,
            x='LapTime',
            y='Driver',
            palette='viridis')
for index, row in fastest_laps_sorted.iterrows():
    plt.text(
        row['LapTime'] + 0.2, 
        index,                       
        f"{row['LapTime']}",          
        va='center'
    )
plt.xlabel('Fastest Lap Time ')
plt.ylabel('Driver')
plt.title('Fastest Lap Per Driver')
plt.tight_layout()
plt.show()
        
```
